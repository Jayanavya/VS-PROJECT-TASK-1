```python
# Step 1: Import necessary libraries
# These are needed for data manipulation (pandas), plotting (matplotlib), and heatmaps (seaborn)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the data
# Since the data is provided as text in the Excel format, we'll create a dictionary and load it into a DataFrame
data = {
    'Transaction_ID': [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011,1012,1013,1014,1015,1016,1017,1018,1019,1020],
    'Date': ['2024-01-05','2024-01-10','2024-01-12','2024-02-05','2024-02-08','2024-02-10','2024-03-15','2024-03-18','2024-03-20','2024-04-01','2024-04-05','2024-04-10','2024-05-01','2024-05-05','2024-05-08','2024-05-10','2024-06-01','2024-06-05','2024-06-08','2024-06-10'],
    'Customer_ID': ['C001','C002','C003','C004','C005','C001','C006','C007','C008','C009','C002','C010','C011','C012','C013','C014','C015','C016','C017','C018'],
    'Product': ['Laptop','Smartphone','Headphones','Tablet','Book','Laptop','Shoes','T-Shirt','Smartwatch','Book','Smartphone','Tablet','Shoes','Headphones','Laptop','T-Shirt','Smartwatch','Book','Smartphone','Tablet'],
    'Category': ['Electronics','Electronics','Electronics','Electronics','Books','Electronics','Clothing','Clothing','Electronics','Books','Electronics','Electronics','Clothing','Electronics','Electronics','Clothing','Electronics','Books','Electronics','Electronics'],
    'Quantity': [1,2,1,1,3,1,2,1,1,2,2,1,1,1,1,3,1,4,1,1],
    'Price': [800,600,100,500,20,800,50,25,200,20,600,500,50,100,800,25,200,20,600,500],
    'Total_Amount': [None,1200,100,500,60,800,100,25,200,40,1200,500,50,100,800,75,200,80,600,500],
    'Payment_Method': ['Credit Card','Cash','PayPal','Debit Card','Credit Card','Credit Card','Cash','PayPal','Debit Card','Credit Card','Cash','Debit Card','Cash','PayPal','Credit Card','Cash','Debit Card','Credit Card','Cash','Debit Card'],
    'Region': ['North','South','West','East','North','North','South','West','East','North','South','East','South','West','North','South','East','North','South','East']
}
df = pd.DataFrame(data)

# Print the original DataFrame to check initial values
print("Original DataFrame:")
print(df)
```

```python
# Step 3: Clean the data
# Handle missing values in Total_Amount by calculating Quantity * Price
df['Total_Amount'] = df['Total_Amount'].fillna(df['Quantity'] * df['Price'])

# Check for duplicates
print("\nNumber of duplicates:", df.duplicated().sum())

# Standardize date format
df['Date'] = pd.to_datetime(df['Date'])

# Add a Month column for grouping
df['Month'] = df['Date'].dt.month

# Print the cleaned DataFrame to verify
print("\nCleaned DataFrame:")
print(df)

# Export to CSV (this will save the file if running locally)
df.to_csv('cleaned_sales_data.csv', index=False)
print("\nCleaned data exported to 'cleaned_sales_data.csv'")
```

```python
# Step 4: Perform Exploratory Data Analysis (EDA)
# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Monthly sales
monthly_sales = df.groupby('Month')['Total_Amount'].sum()
print("\nMonthly Sales:")
print(monthly_sales)

# Category sales
category_sales = df.groupby('Category')['Total_Amount'].sum()
print("\nCategory Sales:")
print(category_sales)

# Product sales
product_sales = df.groupby('Product')['Total_Amount'].sum().sort_values(ascending=False)
print("\nProduct Sales (sorted):")
print(product_sales)

# Repeat customers
repeat_customers = df['Customer_ID'].value_counts()[df['Customer_ID'].value_counts() > 1]
print("\nRepeat Customers:")
print(repeat_customers)

# Correlation matrix
corr_matrix = df[['Quantity', 'Price', 'Total_Amount']].corr()
print("\nCorrelation Matrix:")
print(corr_matrix)
```

```python
# Step 5: Generate Visualizations
# These will display the plots when run, and save them as PNG files

import matplotlib.pyplot as plt
import pandas as pd
# Assuming df from cleaning code
monthly_sales = df.groupby('Month')['Total_Amount'].sum()
plt.figure(figsize=(8,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='blue')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.savefig('sales_over_time.png')  # Save as image
plt.show()

# Visualization 2: Sales by Product Category (Bar Chart)
plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color=['green', 'orange', 'red'])
plt.title('Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.savefig('sales_by_category.png')
plt.show()
print("\nSaved 'sales_by_category.png'")

# Visualization 3: Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')
plt.show()
print("\nSaved 'correlation_heatmap.png'")
```
